```python
"""
=============================================================================
Driver Drowsiness Detection Using CNN + LSTM/GRU (PyTorch)
=============================================================================
Dataset  : kagglehub  talhabhatti7262/drivers-drowsiness-detection
Model    : ResNet50 CNN (feature extractor) + GRU temporal model
Task     : Binary classification - 0 = Alert, 1 = Drowsy
=============================================================================

Install dependencies before running:
    pip install torch torchvision kagglehub opencv-python-headless \
                scikit-learn matplotlib seaborn tqdm Pillow

Run:
    python drowsiness_detection.py
=============================================================================
"""

# ─────────────────────────────────────────────────────────────────────────────
# 0.  Imports & Global Config
# ─────────────────────────────────────────────────────────────────────────────
import os
import time
import random
import warnings
import glob
from pathlib import Path
from collections import defaultdict

import cv2
import numpy as np
from PIL import Image
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report,
)
from sklearn.model_selection import train_test_split
from tqdm import tqdm

warnings.filterwarnings("ignore")

# ── Hyperparameters ──────────────────────────────────────────────────────────
CONFIG = {
    # Data
    "img_size":        224,
    "time_steps":      16,       # frames per sequence
    "frame_stride":    2,        # sample every N-th frame from a clip
    # Model
    "cnn_backbone":    "resnet50",
    "feature_dim":     512,      # CNN output dim (after projection)
    "hidden_dim":      256,      # GRU hidden size
    "num_layers":      2,        # GRU layers
    "dropout":         0.4,
    # Training
    "batch_size":      8,
    "epochs":          1,
    "learning_rate":   1e-4,
    "weight_decay":    1e-5,
    "patience":        5,        # early-stopping patience
    # Deployment
    "alert_threshold": 0.7,
    "device":          "cuda" if torch.cuda.is_available() else "cpu",
    # Paths
    "output_dir":      "outputs",
}

os.makedirs(CONFIG["output_dir"], exist_ok=True)
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

print(f"[Config] Device : {CONFIG['device']}")
print(f"[Config] Hyperparameters:\n"
      f"         batch_size={CONFIG['batch_size']}, "
      f"epochs={CONFIG['epochs']}, "
      f"lr={CONFIG['learning_rate']}, "
      f"time_steps={CONFIG['time_steps']}, "
      f"hidden_dim={CONFIG['hidden_dim']}, "
      f"dropout={CONFIG['dropout']}")


# ─────────────────────────────────────────────────────────────────────────────
# 1.  Dataset Download
# ─────────────────────────────────────────────────────────────────────────────
def download_dataset() -> str:
    """
    Download the Kaggle drowsiness dataset via kagglehub.
    Returns the local root path.

    Requires ~/.kaggle/kaggle.json  OR  env vars
    KAGGLE_USERNAME / KAGGLE_KEY to be set.
    """
    try:
        import kagglehub
        path = kagglehub.dataset_download("talhabhatti7262/drivers-drowsiness-detection")
        print(f"[Dataset] Downloaded to: {path}")
        return path
    except Exception as exc:
        print(f"[Dataset] KaggleHub download failed: {exc}")
        print("[Dataset] Falling back to synthetic demo data ...")
        return None


# ─────────────────────────────────────────────────────────────────────────────
# 2.  Preprocessing Pipeline
# ─────────────────────────────────────────────────────────────────────────────
# Haar cascades shipped with OpenCV - no extra download needed
FACE_CASCADE = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

FRAME_TRANSFORM = transforms.Compose([
    transforms.Resize((CONFIG["img_size"], CONFIG["img_size"])),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],   # ImageNet stats
                         std=[0.229, 0.224, 0.225]),
])


def detect_and_crop_face(frame_bgr: np.ndarray) -> np.ndarray:
    """
    Detect the largest face in a BGR frame and return a cropped RGB patch.
    Falls back to the full frame if no face is detected.
    """
    gray = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2GRAY)
    faces = FACE_CASCADE.detectMultiScale(
        gray, scaleFactor=1.1, minNeighbors=5, minSize=(48, 48)
    )
    if len(faces) == 0:
        # No face detected -> use full frame
        return cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    # Pick the largest face
    x, y, w, h = max(faces, key=lambda r: r[2] * r[3])
    pad = int(0.10 * max(w, h))            # 10 % padding around face
    x1 = max(0, x - pad)
    y1 = max(0, y - pad)
    x2 = min(frame_bgr.shape[1], x + w + pad)
    y2 = min(frame_bgr.shape[0], y + h + pad)
    crop = frame_bgr[y1:y2, x1:x2]
    return cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)


def extract_frames_from_video(
    video_path: str,
    n_frames: int = CONFIG["time_steps"],
    stride: int = CONFIG["frame_stride"],
) -> list[np.ndarray]:
    """
    Extract `n_frames` evenly-spaced frames from a video file.
    Returns a list of preprocessed RGB numpy arrays (H×W×3, uint8).
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return []

    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total <= 0:
        cap.release()
        return []

    # Sample indices spread across the clip
    indices = np.linspace(0, max(0, total - 1), n_frames * stride, dtype=int)[::stride]
    indices = indices[:n_frames]

    frames = []
    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, int(idx))
        ret, frame = cap.read()
        if not ret:
            break
        face_crop = detect_and_crop_face(frame)
        frames.append(face_crop)

    cap.release()

    # Pad with last frame if we got fewer than n_frames
    while len(frames) < n_frames and len(frames) > 0:
        frames.append(frames[-1])

    return frames[:n_frames]


def preprocess_frame(frame_rgb: np.ndarray) -> torch.Tensor:
    """
    Convert a raw RGB numpy array -> normalised tensor [3, H, W].
    """
    img = Image.fromarray(frame_rgb)
    return FRAME_TRANSFORM(img)


def preprocess_sequence(frames: list[np.ndarray]) -> torch.Tensor:
    """
    Convert a list of RGB frames -> tensor [T, 3, H, W].
    """
    tensors = [preprocess_frame(f) for f in frames]
    return torch.stack(tensors)          # (T, 3, 224, 224)


# ─────────────────────────────────────────────────────────────────────────────
# 3.  Dataset Loader
# ─────────────────────────────────────────────────────────────────────────────
def build_file_label_list(dataset_root: str):
    """
    Walk `dataset_root` and collect (file_path, label) pairs.

    Expected layout (flexible):
        dataset_root/
          alert/        ← label 0 (or any folder whose name contains 'alert'/'awake')
          drowsy/       ← label 1 (or 'drowsy'/'sleepy'/'yawn')
          ...

    Videos: .mp4 / .avi / .mov / .mkv
    Images: .jpg / .jpeg / .png  (treated as single-frame sequences)
    """
    VIDEO_EXT = {".mp4", ".avi", ".mov", ".mkv"}
    IMAGE_EXT = {".jpg", ".jpeg", ".png", ".bmp"}

    items = []   # list of (path, label, is_video)

    ALERT_KEYS  = {"alert", "awake", "non_drowsy", "nondrowsy", "active"}
    DROWSY_KEYS = {"drowsy", "sleep", "sleepy", "yawn", "fatigue", "tired"}

    for root, dirs, files in os.walk(dataset_root):
        folder_name = Path(root).name.lower()

        # Determine label from folder name
        if any(k in folder_name for k in ALERT_KEYS):
            label = 0
        elif any(k in folder_name for k in DROWSY_KEYS):
            label = 1
        else:
            continue   # skip unlabelled folders

        for fname in files:
            ext = Path(fname).suffix.lower()
            fpath = os.path.join(root, fname)
            if ext in VIDEO_EXT:
                items.append((fpath, label, True))
            elif ext in IMAGE_EXT:
                items.append((fpath, label, False))

    random.shuffle(items)
    print(f"[Dataset] Found {len(items)} files  "
          f"(alert={sum(1 for _,l,_ in items if l==0)}, "
          f"drowsy={sum(1 for _,l,_ in items if l==1)})")
    return items


class DrowsinessSequenceDataset(Dataset):
    """
    Loads (T, 3, 224, 224) tensors from video clips or image folders.
    For image files: the same image is repeated T times (single-frame clip).
    """

    def __init__(self, items: list, time_steps: int = CONFIG["time_steps"]):
        self.items      = items
        self.time_steps = time_steps

    def __len__(self):
        return len(self.items)

    def __getitem__(self, idx):
        path, label, is_video = self.items[idx]

        if is_video:
            frames = extract_frames_from_video(path, self.time_steps)
        else:
            # Single image -> replicate to form a trivial sequence
            img = cv2.imread(path)
            if img is None:
                frames = []
            else:
                face = detect_and_crop_face(img)
                frames = [face] * self.time_steps

        # Fallback: black frames
        if len(frames) == 0:
            frames = [np.zeros((CONFIG["img_size"], CONFIG["img_size"], 3),
                               dtype=np.uint8)] * self.time_steps

        seq_tensor = preprocess_sequence(frames)   # (T, 3, H, W)
        return seq_tensor, torch.tensor(label, dtype=torch.float32)


def make_dataloaders(dataset_root: str):
    """
    Build train / val / test DataLoaders from a dataset root directory.
    Falls back to synthetic data if root is None or empty.
    """
    if dataset_root and os.path.isdir(dataset_root):
        items = build_file_label_list(dataset_root)
    else:
        items = []

    if len(items) < 10:
        print("[Dataset] Using synthetic data for demonstration ...")
        items = _make_synthetic_items(n=200)

    train_items, test_items = train_test_split(
        items, test_size=0.15, random_state=42,
        stratify=[l for _, l, _ in items]
    )
    train_items, val_items = train_test_split(
        train_items, test_size=0.15, random_state=42,
        stratify=[l for _, l, _ in train_items]
    )

    def make_loader(split_items, shuffle):
        ds = DrowsinessSequenceDataset(split_items)
        return DataLoader(
            ds, batch_size=CONFIG["batch_size"],
            shuffle=shuffle, num_workers=0, pin_memory=False,
        )

    loaders = {
        "train": make_loader(train_items, shuffle=True),
        "val":   make_loader(val_items,   shuffle=False),
        "test":  make_loader(test_items,  shuffle=False),
    }
    print(f"[Dataset] Split -> train={len(train_items)}, "
          f"val={len(val_items)}, test={len(test_items)}")
    return loaders


def _make_synthetic_items(n: int = 200):
    """
    Generate synthetic (random-image, label) pairs for offline testing.
    Images are tiny solid-colour PNGs written to /tmp.
    """
    tmp = Path("/tmp/synthetic_drowsy")
    for cls, lbl in [("alert", 0), ("drowsy", 1)]:
        (tmp / cls).mkdir(parents=True, exist_ok=True)

    items = []
    for i in range(n):
        label = i % 2
        cls   = "alert" if label == 0 else "drowsy"
        fpath = str(tmp / cls / f"frame_{i:04d}.jpg")
        if not os.path.exists(fpath):
            colour = (100, 200, 100) if label == 0 else (200, 80, 80)
            img    = np.full((64, 64, 3), colour, dtype=np.uint8)
            cv2.imwrite(fpath, img)
        items.append((fpath, label, False))

    return items


# ─────────────────────────────────────────────────────────────────────────────
# 4.  Model Architecture: CNN (ResNet50) + GRU
# ─────────────────────────────────────────────────────────────────────────────
class CNNFeatureExtractor(nn.Module):
    """
    ResNet50 backbone with the final FC layer replaced by a projection head.
    The convolutional body is frozen; only the projection head is trained.

    Output shape: (batch_size, feature_dim)
    """

    def __init__(self, feature_dim: int = CONFIG["feature_dim"]):
        super().__init__()

        backbone = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)

        # Freeze all conv layers
        for param in backbone.parameters():
            param.requires_grad = False

        in_features = backbone.fc.in_features          # 2048

        # Replace final FC with a trainable projection head
        backbone.fc = nn.Sequential(
            nn.Linear(in_features, feature_dim),
            nn.BatchNorm1d(feature_dim),
            nn.ReLU(inplace=True),
            nn.Dropout(p=CONFIG["dropout"]),
        )

        self.backbone   = backbone
        self.feature_dim = feature_dim

    def train(self, mode: bool = True):
        super().train(mode)
        # Keep frozen backbone in eval mode to prevent BatchNorm corruption
        self.backbone.eval()
        self.backbone.fc.train(mode)
        return self

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """x: (B, 3, H, W)  ->  (B, feature_dim)"""
        return self.backbone(x)


class DrowsinessGRU(nn.Module):
    """
    GRU temporal model that ingests per-frame CNN features.

    Forward pass:
        x  : (B, T, feature_dim)  - sequence of CNN embeddings
        out: (B, 1)               - drowsiness probability (after sigmoid)
    """

    def __init__(
        self,
        feature_dim: int = CONFIG["feature_dim"],
        hidden_dim:  int = CONFIG["hidden_dim"],
        num_layers:  int = CONFIG["num_layers"],
        dropout:     float = CONFIG["dropout"],
    ):
        super().__init__()

        self.gru = nn.GRU(
            input_size=feature_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0,
            bidirectional=False,
        )
        self.attention = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.Tanh(),
            nn.Linear(hidden_dim // 2, 1),
        )
        self.classifier = nn.Sequential(
            nn.Dropout(p=dropout),
            nn.Linear(hidden_dim, 64),
            nn.ReLU(inplace=True),
            nn.Linear(64, 1),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        """x: (B, T, feature_dim)"""
        gru_out, _ = self.gru(x)           # (B, T, hidden_dim)

        # Soft attention over time steps
        attn_scores = self.attention(gru_out)           # (B, T, 1)
        attn_weights = torch.softmax(attn_scores, dim=1)
        context = (attn_weights * gru_out).sum(dim=1)  # (B, hidden_dim)

        logit = self.classifier(context)               # (B, 1)
        return torch.sigmoid(logit)


class DrowsinessDetector(nn.Module):
    """
    End-to-end model: CNN (per frame) + GRU (across time).

    Input : (B, T, 3, H, W)
    Output: (B, 1)  - probability of drowsiness [0, 1]
    """

    def __init__(self):
        super().__init__()
        self.cnn = CNNFeatureExtractor(feature_dim=CONFIG["feature_dim"])
        self.gru = DrowsinessGRU(
            feature_dim=CONFIG["feature_dim"],
            hidden_dim=CONFIG["hidden_dim"],
            num_layers=CONFIG["num_layers"],
            dropout=CONFIG["dropout"],
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        B, T, C, H, W = x.shape

        # Merge batch & time dims -> run CNN once over all frames
        x_flat     = x.view(B * T, C, H, W)          # (B*T, 3, H, W)
        feats_flat = self.cnn(x_flat)                 # (B*T, feature_dim)
        feats      = feats_flat.view(B, T, -1)        # (B, T, feature_dim)

        prob = self.gru(feats)                        # (B, 1)
        return prob


def model_summary(model: nn.Module, device: str):
    total  = sum(p.numel() for p in model.parameters())
    train  = sum(p.numel() for p in model.parameters() if p.requires_grad)
    frozen = total - train
    print(f"\n[Model] Architecture: CNN (ResNet50) + GRU")
    print(f"        Total params   : {total:,}")
    print(f"        Trainable      : {train:,}")
    print(f"        Frozen (CNN)   : {frozen:,}")
    print(f"        Device         : {device}\n")


# ─────────────────────────────────────────────────────────────────────────────
# 5.  Training Loop
# ─────────────────────────────────────────────────────────────────────────────
def train_one_epoch(
    model: nn.Module,
    loader: DataLoader,
    criterion: nn.Module,
    optimizer: optim.Optimizer,
    device: str,
) -> tuple[float, float]:
    """Returns (avg_loss, accuracy)."""
    model.train()
    total_loss, correct, total = 0.0, 0, 0

    for seqs, labels in tqdm(loader, desc="  Train", leave=False):
        seqs   = seqs.to(device)          # (B, T, 3, 224, 224)
        labels = labels.to(device)        # (B,)

        optimizer.zero_grad()
        probs = model(seqs).squeeze(1)    # (B,)
        loss  = criterion(probs, labels)
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        total_loss += loss.item() * seqs.size(0)
        preds = (probs >= 0.5).long()
        correct += (preds == labels.long()).sum().item()
        total   += seqs.size(0)

    return total_loss / total, correct / total


@torch.no_grad()
def evaluate(
    model: nn.Module,
    loader: DataLoader,
    criterion: nn.Module,
    device: str,
) -> tuple[float, float, np.ndarray, np.ndarray]:
    """Returns (avg_loss, accuracy, all_labels, all_probs)."""
    model.eval()
    total_loss, correct, total = 0.0, 0, 0
    all_labels, all_probs = [], []

    for seqs, labels in tqdm(loader, desc="  Eval ", leave=False):
        seqs   = seqs.to(device)
        labels = labels.to(device)

        probs = model(seqs).squeeze(1)
        loss  = criterion(probs, labels)

        total_loss += loss.item() * seqs.size(0)
        preds = (probs >= 0.5).long()
        correct += (preds == labels.long()).sum().item()
        total   += seqs.size(0)

        all_labels.extend(labels.cpu().numpy())
        all_probs.extend(probs.cpu().numpy())

    return (
        total_loss / total,
        correct / total,
        np.array(all_labels),
        np.array(all_probs),
    )


def train(model: nn.Module, loaders: dict, device: str) -> dict:
    """
    Full training loop with:
    - Binary Cross-Entropy loss
    - AdamW optimiser + cosine LR scheduler
    - Early stopping
    Returns history dict.
    """
    criterion = nn.BCELoss()
    optimizer = optim.AdamW(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=CONFIG["learning_rate"],
        weight_decay=CONFIG["weight_decay"],
    )
    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=CONFIG["epochs"], eta_min=1e-6
    )

    history = defaultdict(list)
    best_val_loss = float("inf")
    patience_counter = 0
    best_state = None

    print(f"\n[Train] Starting - {CONFIG['epochs']} epochs, "
          f"patience={CONFIG['patience']}")
    print("-" * 60)

    for epoch in range(1, CONFIG["epochs"] + 1):
        t0 = time.time()

        tr_loss, tr_acc = train_one_epoch(
            model, loaders["train"], criterion, optimizer, device
        )
        va_loss, va_acc, _, _ = evaluate(
            model, loaders["val"], criterion, device
        )
        scheduler.step()

        history["tr_loss"].append(tr_loss)
        history["tr_acc"].append(tr_acc)
        history["va_loss"].append(va_loss)
        history["va_acc"].append(va_acc)

        elapsed = time.time() - t0
        print(f"  Epoch {epoch:3d}/{CONFIG['epochs']}  "
              f"train_loss={tr_loss:.4f}  train_acc={tr_acc:.4f}  "
              f"val_loss={va_loss:.4f}  val_acc={va_acc:.4f}  "
              f"({elapsed:.1f}s)")

        # Early stopping
        if va_loss < best_val_loss - 1e-4:
            best_val_loss    = va_loss
            patience_counter = 0
            best_state       = {k: v.clone() for k, v in model.state_dict().items()}
        else:
            patience_counter += 1
            if patience_counter >= CONFIG["patience"]:
                print(f"  [EarlyStopping] No improvement for "
                      f"{CONFIG['patience']} epochs - stopping.")
                break

    # Restore best weights
    if best_state is not None:
        model.load_state_dict(best_state)
        print("[Train] Best weights restored.")

    # Save checkpoint
    ckpt_path = os.path.join(CONFIG["output_dir"], "best_model.pt")
    torch.save(model.state_dict(), ckpt_path)
    print(f"[Train] Checkpoint saved -> {ckpt_path}")

    return dict(history)


# ─────────────────────────────────────────────────────────────────────────────
# 6.  Evaluation & Visualisation
# ─────────────────────────────────────────────────────────────────────────────
def full_evaluation(
    model: nn.Module,
    loader: DataLoader,
    device: str,
    threshold: float = CONFIG["alert_threshold"],
) -> dict:
    """
    Run the test set through the model and compute all metrics.
    Saves confusion matrix and training curves as PNG files.
    """
    criterion = nn.BCELoss()
    _, _, y_true, y_prob = evaluate(model, loader, criterion, device)
    y_pred = (y_prob >= threshold).astype(int)

    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)
    cm   = confusion_matrix(y_true, y_pred)

    print("\n" + "=" * 60)
    print("  EVALUATION RESULTS")
    print("=" * 60)
    print(f"  Threshold   : {threshold}")
    print(f"  Accuracy    : {acc:.4f}")
    print(f"  Precision   : {prec:.4f}")
    print(f"  Recall      : {rec:.4f}")
    print(f"  F1-score    : {f1:.4f}")
    print("\n  Classification Report:")
    print(classification_report(y_true, y_pred,
                                target_names=["Alert", "Drowsy"],
                                zero_division=0))

    # ── Confusion Matrix ─────────────────────────────────────────────────────
    fig, ax = plt.subplots(figsize=(5, 4))
    sns.heatmap(
        cm, annot=True, fmt="d", cmap="Blues",
        xticklabels=["Pred Alert", "Pred Drowsy"],
        yticklabels=["True Alert", "True Drowsy"],
        ax=ax, linewidths=0.5,
    )
    ax.set_title("Confusion Matrix", fontsize=13, pad=10)
    ax.set_ylabel("Actual")
    ax.set_xlabel("Predicted")
    plt.tight_layout()
    cm_path = os.path.join(CONFIG["output_dir"], "confusion_matrix.png")
    plt.savefig(cm_path, dpi=150)
    plt.close()
    print(f"\n[Eval] Confusion matrix saved -> {cm_path}")

    return {
        "accuracy": acc, "precision": prec,
        "recall": rec, "f1": f1,
        "confusion_matrix": cm,
        "y_true": y_true, "y_prob": y_prob,
    }


def plot_training_curves(history: dict):
    """Save loss & accuracy curves."""
    epochs = range(1, len(history["tr_loss"]) + 1)

    fig, axes = plt.subplots(1, 2, figsize=(10, 4))

    axes[0].plot(epochs, history["tr_loss"], label="Train Loss",  color="#3B8BD4")
    axes[0].plot(epochs, history["va_loss"], label="Val Loss",    color="#E24B4A", linestyle="--")
    axes[0].set_title("Loss")
    axes[0].set_xlabel("Epoch")
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    axes[1].plot(epochs, history["tr_acc"], label="Train Acc",   color="#1D9E75")
    axes[1].plot(epochs, history["va_acc"], label="Val Acc",     color="#EF9F27", linestyle="--")
    axes[1].set_title("Accuracy")
    axes[1].set_xlabel("Epoch")
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.suptitle("Training Curves - CNN + GRU Drowsiness Detector", fontsize=12)
    plt.tight_layout()
    curve_path = os.path.join(CONFIG["output_dir"], "training_curves.png")
    plt.savefig(curve_path, dpi=150)
    plt.close()
    print(f"[Eval] Training curves saved -> {curve_path}")


# ─────────────────────────────────────────────────────────────────────────────
# 7.  Inference Pipeline & Latency Analysis
# ─────────────────────────────────────────────────────────────────────────────
@torch.no_grad()
def infer_single_sequence(
    model: nn.Module,
    frames: list[np.ndarray],
    device: str,
    threshold: float = CONFIG["alert_threshold"],
) -> dict:
    """
    Run drowsiness detection on a list of raw RGB frames.

    Returns a dict with:
        probability  - float in [0, 1]
        prediction   - "Drowsy" | "Alert"
        alert        - bool
        latency_ms   - inference time in milliseconds
    """
    model.eval()

    # Pad / truncate to exactly time_steps frames
    T = CONFIG["time_steps"]
    if len(frames) < T:
        frames = frames + [frames[-1]] * (T - len(frames))
    frames = frames[:T]

    seq = preprocess_sequence(frames)              # (T, 3, H, W)
    seq = seq.unsqueeze(0).to(device)              # (1, T, 3, H, W)

    start   = time.perf_counter()
    prob    = model(seq).item()
    latency = (time.perf_counter() - start) * 1000  # ms

    prediction = "Drowsy" if prob >= threshold else "Alert"

    return {
        "probability":  round(prob, 4),
        "prediction":   prediction,
        "alert":        prob >= threshold,
        "latency_ms":   round(latency, 2),
    }


def benchmark_latency(
    model: nn.Module,
    device: str,
    n_runs: int = 50,
    warmup: int = 5,
) -> dict:
    """
    Measure inference latency over `n_runs` random batches.
    Returns stats: mean, std, p50, p95, p99 (all in ms).
    """
    model.eval()
    B  = 1
    T  = CONFIG["time_steps"]
    C, H, W = 3, CONFIG["img_size"], CONFIG["img_size"]

    latencies = []

    with torch.no_grad():
        dummy = torch.randn(B, T, C, H, W, device=device)

        # Warm-up
        for _ in range(warmup):
            _ = model(dummy)

        for _ in range(n_runs):
            if device == "cuda":
                torch.cuda.synchronize()
            t0 = time.perf_counter()
            _ = model(dummy)
            if device == "cuda":
                torch.cuda.synchronize()
            latencies.append((time.perf_counter() - t0) * 1000)

    arr = np.array(latencies)
    stats = {
        "mean_ms": round(float(arr.mean()), 2),
        "std_ms":  round(float(arr.std()),  2),
        "p50_ms":  round(float(np.percentile(arr, 50)), 2),
        "p95_ms":  round(float(np.percentile(arr, 95)), 2),
        "p99_ms":  round(float(np.percentile(arr, 99)), 2),
        "n_runs":  n_runs,
    }

    fps_estimate = 1000.0 / stats["mean_ms"]
    print("\n[Latency] Inference Benchmark")
    print(f"          Runs    : {n_runs}")
    print(f"          Mean    : {stats['mean_ms']} ms")
    print(f"          Std     : {stats['std_ms']} ms")
    print(f"          P50     : {stats['p50_ms']} ms")
    print(f"          P95     : {stats['p95_ms']} ms")
    print(f"          P99     : {stats['p99_ms']} ms")
    print(f"          Est FPS : {fps_estimate:.1f} "
          f"({'feasible' if fps_estimate >= 5 else 'too slow'} for real-time @ 5 fps)")

    # Plot latency distribution
    fig, ax = plt.subplots(figsize=(7, 3))
    ax.hist(arr, bins=20, color="#3B8BD4", edgecolor="white", alpha=0.85)
    ax.axvline(stats["mean_ms"], color="#E24B4A", linestyle="--",
               label=f"Mean = {stats['mean_ms']} ms")
    ax.axvline(stats["p95_ms"],  color="#EF9F27", linestyle=":",
               label=f"P95  = {stats['p95_ms']} ms")
    ax.set_xlabel("Latency (ms)")
    ax.set_ylabel("Count")
    ax.set_title("Inference Latency Distribution")
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    lat_path = os.path.join(CONFIG["output_dir"], "latency_analysis.png")
    plt.savefig(lat_path, dpi=150)
    plt.close()
    print(f"[Latency] Distribution saved -> {lat_path}")

    return stats


# ─────────────────────────────────────────────────────────────────────────────
# 8.  Alert Threshold Configuration
# ─────────────────────────────────────────────────────────────────────────────
def analyze_threshold(y_true: np.ndarray, y_prob: np.ndarray):
    """
    Sweep threshold values and plot Precision / Recall / F1 vs threshold.
    Also prints the optimal threshold (max F1).
    """
    thresholds = np.linspace(0.1, 0.95, 50)
    precisions, recalls, f1s = [], [], []

    for t in thresholds:
        y_pred = (y_prob >= t).astype(int)
        precisions.append(precision_score(y_true, y_pred, zero_division=0))
        recalls.append(recall_score(y_true, y_pred, zero_division=0))
        f1s.append(f1_score(y_true, y_pred, zero_division=0))

    best_idx = int(np.argmax(f1s))
    best_t   = thresholds[best_idx]

    print(f"\n[Threshold] Optimal threshold (max F1): {best_t:.2f}  "
          f"->  F1={f1s[best_idx]:.4f}, "
          f"Prec={precisions[best_idx]:.4f}, "
          f"Rec={recalls[best_idx]:.4f}")

    fig, ax = plt.subplots(figsize=(7, 4))
    ax.plot(thresholds, precisions, label="Precision", color="#3B8BD4")
    ax.plot(thresholds, recalls,    label="Recall",    color="#1D9E75")
    ax.plot(thresholds, f1s,        label="F1-score",  color="#E24B4A", linewidth=2)
    ax.axvline(best_t, color="#EF9F27", linestyle="--",
               label=f"Optimal = {best_t:.2f}")
    ax.axvline(CONFIG["alert_threshold"], color="#888", linestyle=":",
               label=f"Configured = {CONFIG['alert_threshold']}")
    ax.set_xlabel("Threshold")
    ax.set_ylabel("Score")
    ax.set_title("Alert Threshold Analysis")
    ax.legend(fontsize=9)
    ax.set_xlim([0.1, 0.95])
    ax.set_ylim([0, 1.05])
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    thr_path = os.path.join(CONFIG["output_dir"], "threshold_analysis.png")
    plt.savefig(thr_path, dpi=150)
    plt.close()
    print(f"[Threshold] Curve saved -> {thr_path}")
    return best_t


# ─────────────────────────────────────────────────────────────────────────────
# 9.  Real-time Webcam Inference with Visual Display
# ─────────────────────────────────────────────────────────────────────────────
def run_webcam_inference(model: nn.Module, device: str, camera_id: int = 0):
    """
    Real-time drowsiness detection from webcam with visual overlay.
    
    - Reads frames from webcam
    - Displays prediction text and confidence on video
    - Press 'q' to quit
    - Shows FPS counter
    """
    print("\n[Webcam] Starting real-time inference...")
    print("         Press 'q' to quit")
    print("-" * 60)
    
    cap = cv2.VideoCapture(camera_id)
    if not cap.isOpened():
        print("[Webcam] ERROR: Could not open camera!")
        return
    
    # Set camera resolution for faster processing
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    
    frame_buffer = []  # Store frames for sequence inference
    fps_counter = 0
    fps_timer = time.time()
    current_fps = 0
    
    model.eval()
    
    try:
        while True:
            ret, frame_bgr = cap.read()
            if not ret:
                print("[Webcam] Failed to read frame")
                break
            
            # Detect and crop face
            face_crop_rgb = detect_and_crop_face(frame_bgr)
            
            # Add to buffer
            frame_buffer.append(face_crop_rgb)
            if len(frame_buffer) > CONFIG["time_steps"]:
                frame_buffer.pop(0)
            
            # Run inference when we have enough frames
            prediction_text = "Loading..."
            confidence = 0.0
            alert_triggered = False
            latency_ms = 0.0
            
            if len(frame_buffer) == CONFIG["time_steps"]:
                result = infer_single_sequence(model, frame_buffer, device)
                prediction_text = result["prediction"]
                confidence = result["probability"]
                alert_triggered = result["alert"]
                latency_ms = result["latency_ms"]
            
            # Draw on original frame
            display_frame = frame_bgr.copy()
            h, w = display_frame.shape[:2]
            
            # ── Background box for text ──────────────────────────────────────
            box_height = 80
            cv2.rectangle(display_frame, (0, 0), (w, box_height), 
                         (0, 0, 0), -1)  # Black background
            
            # ── Prediction text (large) ──────────────────────────────────────
            color = (0, 0, 255) if alert_triggered else (0, 255, 0)  # Red or Green
            text = f"Prediction: {prediction_text}"
            cv2.putText(display_frame, text, (20, 50),
                       cv2.FONT_HERSHEY_SIMPLEX, 1.5, color, 3)
            
            # ── Confidence + Latency (small) ─────────────────────────────────
            info_text = f"Confidence: {confidence:.2%}  |  Latency: {latency_ms} ms"
            cv2.putText(display_frame, info_text, (20, 75),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200, 200, 200), 2)
            
            # ── FPS counter ──────────────────────────────────────────────────
            fps_counter += 1
            if time.time() - fps_timer > 1.0:
                current_fps = fps_counter
                fps_counter = 0
                fps_timer = time.time()
            
            fps_text = f"FPS: {current_fps}"
            cv2.putText(display_frame, fps_text, (w - 150, 40),
                       cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 255), 2)
            
            # ── Alert indicator (if drowsy) ──────────────────────────────────
            if alert_triggered:
                cv2.rectangle(display_frame, (5, 5), (35, 35), (0, 0, 255), -1)
                cv2.circle(display_frame, (20, 20), 8, (0, 255, 255), 3)
            
            # Display frame
            cv2.imshow("Drowsiness Detection", display_frame)
            
            # Check for quit
            key = cv2.waitKey(1) & 0xFF
            if key == ord('q'):
                print("[Webcam] Quit requested by user")
                break
    
    except KeyboardInterrupt:
        print("\n[Webcam] Interrupted by user")
    finally:
        cap.release()
        cv2.destroyAllWindows()
        print("[Webcam] Camera released and display closed")


def run_demo_inference(model: nn.Module, device: str):
    """
    Simulate real-time inference on 5 random sequences and print results.
    In a live deployment, `frames` would come from a webcam stream.
    """
    print("\n[Demo] Simulated real-time inference on 5 random sequences")
    print("-" * 60)

    for i in range(5):
        # In production: replace with actual frames from cv2.VideoCapture
        frames = [
            np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8)
            for _ in range(CONFIG["time_steps"])
        ]
        result = infer_single_sequence(model, frames, device)

        alert_str = "!  ALERT TRIGGERED" if result["alert"] else "   OK"
        print(
            f"  Sample {i+1}:  prob={result['probability']:.4f}  "
            f"pred={result['prediction']:7s}  "
            f"latency={result['latency_ms']} ms  {alert_str}"
        )
    print("-" * 60)


# ─────────────────────────────────────────────────────────────────────────────
# 10.  Main
# ─────────────────────────────────────────────────────────────────────────────
def main():
    device = CONFIG["device"]

    # ── Step 1: Download dataset ─────────────────────────────────────────────
    print("\n" + "=" * 60)
    print("  STEP 1 - Dataset")
    print("=" * 60)
    dataset_root = download_dataset()

    # ── Step 2: Build data loaders ───────────────────────────────────────────
    print("\n" + "=" * 60)
    print("  STEP 2 - Preprocessing & DataLoaders")
    print("=" * 60)
    loaders = make_dataloaders(dataset_root)

    # ── Step 3: Build model ──────────────────────────────────────────────────
    print("\n" + "=" * 60)
    print("  STEP 3 - Model Architecture")
    print("=" * 60)
    model = DrowsinessDetector().to(device)
    model_summary(model, device)

    # ── Step 4: Train ────────────────────────────────────────────────────────
    print("\n" + "=" * 60)
    print("  STEP 4 - Training")
    print("=" * 60)
    history = train(model, loaders, device)
    plot_training_curves(history)

    # ── Step 5: Evaluate ─────────────────────────────────────────────────────
    print("\n" + "=" * 60)
    print("  STEP 5 - Evaluation")
    print("=" * 60)
    results = full_evaluation(model, loaders["test"], device)
    analyze_threshold(results["y_true"], results["y_prob"])

    # ── Step 6: Latency benchmark ────────────────────────────────────────────
    print("\n" + "=" * 60)
    print("  STEP 6 - Latency Analysis")
    print("=" * 60)
    benchmark_latency(model, device, n_runs=50)

    # ── Step 7: Real-time Webcam Inference ──────────────────────────────────
    print("\n" + "=" * 60)
    print("  STEP 7 - Real-time Webcam Inference")
    print("=" * 60)
    run_webcam_inference(model, device)

    # ── Summary ──────────────────────────────────────────────────────────────
    print("\n" + "=" * 60)
    print("  SUMMARY")
    print("=" * 60)
    print(f"  Model      : CNN (ResNet50 backbone) + GRU")
    print(f"  Batch size : {CONFIG['batch_size']}")
    print(f"  Epochs     : {CONFIG['epochs']}")
    print(f"  LR         : {CONFIG['learning_rate']}")
    print(f"  Time steps : {CONFIG['time_steps']}")
    print(f"  Feature dim: {CONFIG['feature_dim']}")
    print(f"  Hidden dim : {CONFIG['hidden_dim']}")
    print(f"  GRU layers : {CONFIG['num_layers']}")
    print(f"  Dropout    : {CONFIG['dropout']}")
    print(f"  Threshold  : {CONFIG['alert_threshold']}")
    print(f"  Accuracy   : {results['accuracy']:.4f}")
    print(f"  F1-score   : {results['f1']:.4f}")
    print(f"  Outputs    : {CONFIG['output_dir']}/")
    print("=" * 60)


if __name__ == "__main__":
    main()
```
